# get_dummies


One Hot Encoding and get_dummies almost equal . Major difference is if you want to reduce (drop_first=True) the column size of the dataset you can use get_dummies.


OHE does'nt add variables names to your dataframe , but get_dummies add varibale names.



Sometimes having more column might overfit the model.



```python
df1=pd.get_dummies(df[['City','Gender']])
df1
```

```python
df1=pd.get_dummies(df[['City','Gender']],drop_first=True)
df1
```

# Ordinal Encoder 

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
import pandas as pd


employee=pd.DataFrame({'Position':['SE','Manager','Team Leader','SSE'],
                      'Project':['A','B','C','D'],
                      'Salary':[25000,85000,71000,48000]})
employee
```

# KNN imputer

```python
df=pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,49000],
                'City':['Bengluru','Delhi','Hyderabad','Bengluru','Hyderabad','Bengluru'],
                'Exp':[1,3,5,6,9,None]})
df
```

KNN imputer will try to find the relation with other columns and impute the data according the relation with other columns.


In this case Age Nan is depending  on the similarity with Fare columns/

```python
from sklearn.impute  import KNNImputer
```

```python
knn_imp=KNNImputer(n_neighbors=2)
knn_imp=pd.DataFrame(knn_imp.fit_transform(df[['Salary','Exp']]))

knn_imp
```

#  Iterative Imputer


This method treat other columns (which does not have nulls as feature and train on them and treat null column as label .finally it will predict the Nan data and  impute . Its just like regression problem .Here null columns is label.

```python
#Before using Iterative Imputer , we need to enable it using below code.

from sklearn.experimental import enable_iterative_imputer
```

```python
#Import Iterative Imputer
```

```python
from sklearn.impute import IterativeImputer
```

```python
import pandas as pd


df=pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,55000],
                'City':['Bengluru','Delhi','Hyderabad','Bengluru','Hyderabad','Bengluru'],
                'Gender':['Male','Female','Female','Male','Female','Male'],
                'Exp':[1,3,5,6,9,None]})
df
```

```python
iter_impute=IterativeImputer()
ite_imp=pd.DataFrame(iter_impute.fit_transform(df[['Salary','Exp']]),columns=['Salary','Exp'])

ite_imp
```
