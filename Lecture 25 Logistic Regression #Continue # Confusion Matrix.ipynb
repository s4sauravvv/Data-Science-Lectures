# Advantage of Logistic Regression


- It is very simple and easy to implement


- The output is more informative than other classification algorithms


- It express the relationship between independent and dependent  variables 



- Very effective with linearly separable data.




# Disadvantage of Logistic Regression



- Not effective with data which are not limearly separable



- Not as powerfull as other classification



- Multiclass classification are much easier to do with other algorithms than losgistic regression.




- It can only predict categorical outcomes 



- Not effective the labels are more than two 




- Logistic Regression is very good for binary classification.
